## PROBA.QA.7
Let $ {\mathbf x} $ be a random variable taking the values -1, 0 and 1 with probabilities 1/3,1/3,1/3 respectively and 
let $$ {\mathbf y}={\mathbf x}^2 $$
What is the probability $ P( {\mathbf x}=1| {\mathbf y}=1) $ ?

Test the result in Python by using Monte Carlo.



### Solution
By using Bayes' theorem

 $$ P( {\mathbf x}=1| {\mathbf y}=1) = \frac{P( {\mathbf y}=1| {\mathbf x}=1)P( {\mathbf x}=1) }{P( {\mathbf y}=1| {\mathbf x}=1)P( {\mathbf x}=1) +P( {\mathbf y}=1| {\mathbf x}=0)P( {\mathbf x}=0)+ P( {\mathbf y}=1| {\mathbf x}=-1)P( {\mathbf x}=-1) }=  \frac{1*1/3}{2/3}=1/2 $$

```python
import numpy as np
R = 50000

X = np.random.choice([-1, 0, 1], size=R, replace=True)
Y = X**2

I = np.where(Y == 1)[0]
print(np.sum(X[I] == 1) / len(I))
```

## PROBA.QA.11

Let us consider three r.v.s $\mathbf x, \mathbf y,\mathbf z$. 

1. Prove that

$$\mbox{Cov}[\mathbf x+ \mathbf y , \mathbf z] = \mbox{Cov}[\mathbf x , \mathbf z]+ 
\mbox{Cov}[\mathbf y , \mathbf z] $$

2. Test the result in Python by using Monte Carlo and supposing that $\mathbf x \sim U(-2,2)$ and $\mathbf y,\mathbf z$ are standard normal.

### Solution
$$ \mbox{Cov}[\mathbf x+ \mathbf y,\mathbf z]=E[\mathbf z(\mathbf x+\mathbf y)]-E[\mathbf x+\mathbf y]*E[\mathbf z]=E[\mathbf z \mathbf x]+E[\mathbf z \mathbf y]-E[\mathbf x]E[\mathbf z]-E[\mathbf y]E[\mathbf z]=\mbox{Cov}[\mathbf z, \mathbf x]+\mbox{Cov}[\mathbf y, \mathbf z]$$

```python
import numpy as np

N = 1000

x = np.random.uniform(-2.0, 2.0, size=N)
y = np.random.normal(loc=0.0, scale=1.0, size=N)
z = np.random.normal(loc=0.0, scale=1.0, size=N)

# Calculate covariance between (x+y) and z, extracting the off-diagonal element from the covariance matrix.
cov_xy_z = np.cov(x + y, z)[0, 1]
# Calculate covariance between x and z.
cov_x_z = np.cov(x, z)[0, 1]
# Calculate covariance between y and z.
cov_y_z = np.cov(y, z)[0, 1]

print(cov_xy_z, ":", cov_x_z + cov_y_z)
```

## PROBA.Q.12
Let $\mathbf x$ and $\mathbf y$ two independent standard random variables. 

Let $\mathbf z=\mathbf x$ and $\mathbf w=\rho \mathbf x +\sqrt{1-\rho^2}\, \mathbf y$ with $-1 \le \rho \le 1$

1. Compute the correlation between $\mathbf z$ and $\mathbf w$.
2. Test the result in Python


### Solution
Since $\mbox{Var}[\mathbf y]=\mbox{Var}[\mathbf x]=\mbox{Var}[\mathbf z]=1$, we have
$\mbox{Var}[\mathbf w]=\rho^2+1-\rho^2=1$ and $\mbox{Cor}[\mathbf z,\mathbf w]=\mbox{Cov}[\mathbf z,\mathbf w]$

Since 
* $E[\mathbf x]=E[\mathbf w]=0$
* $ E[\mathbf x^2]=\mbox{Var}[\mathbf x] - E[\mathbf x]^2= 1$
* $ E[\mathbf  x \mathbf y]=0$ ($\mathbf  x$ and $\mathbf  y$ are independent) we have 

$$\mbox{Cor}[\mathbf z, \mathbf w]=\mbox{Cov}[\mathbf z,\mathbf w]=
E[\mathbf z \mathbf w]=E[\rho \mathbf x^2+\sqrt{1-\rho^2} \, \mathbf x \mathbf y]=\rho E[ \mathbf x^2]= \rho$$



```python
import numpy as np
N = 100000
for rho in np.arange(-1, 1.1, 0.2):
    X = np.random.normal(size=N)
    Y = np.random.normal(size=N)
    Z = X
    W = rho * X + np.sqrt(1 - rho**2) * Y
    print(np.round(rho,2), ":", np.round(np.corrcoef(Z, W)[0, 1],2))
```
