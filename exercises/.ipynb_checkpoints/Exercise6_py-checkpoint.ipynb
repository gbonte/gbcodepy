## Question
Consider a binary classification task where the input ${\mathbf x} \in {\mathbb R}^2$ is bivariate and the categorical output variable ${\mathbf y}$  may take
two values: $0$ (associated to red) and $1$ (associated to green).

Suppose that the a-priori probability is 
$p({\mathbf y}=1)=0.4$ and that the inverse (or class-conditional) distributions
are 

* green/cross class : mixture of three Gaussians 
$$ p(x|{\mathbf y}=1)=\sum_{i=1}^3 w_i \mathcal{N}(\mu_{1i},\Sigma)$$
where in $\mu_{11}=[1,1]^T, \mu_{12}=[-1,-1], \mu_{13}=[3,-3]^T$, 
and $w_1=0.2$, $w_2=0.3$.
* red/circel class: bivariate Gaussian
$p(x | {\mathbf y}=0)={\mathcal N}(\mu_0,\Sigma)$
where $\mu_0=[0,0]^T$

The matrix $\Sigma$ is a  diagonal identity matrix.

The student should

* by using the R function ${\tt rmvnorm}$, sample a dataset of $N=1000$ input/output observations according to the conditional distribution described above,
* visualise in a 2D graph the dataset by using the appropriate colors,
* plot the ROC curves of the following classifiers

  1. linear regression coding the two classes by 0 and 1,
  2. Linear Discriminant Analysis where $\sigma^2=1$,
  3. Naive Bayes where the univariate conditional distributions are Gaussian,
  4. k Nearest Neighbour with $k=3,5,10$.

  The classifiers should be trained and tested on the same training set.

* Choose the best classifier on the basis of the ROC curves above.

\newpage

## Data generation

The sampling of data from a mixture of gaussians requires two steps: 
first the sampling of the component (using the distribution characterized
by the weights) and then the sampling of the associated gaussian distribution.
Equivalently you may for each $i$th component draw a number of samples
equal to $w_i N$.

```python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt

N = 1000

p1 = 0.4

w1 = 0.2
w2 = 0.3

mu11 = np.array([1, 1])
mu12 = np.array([-1, -1])
mu13 = np.array([3, -3])
Sigma1 = np.eye(2)
N1 = int(N * p1)

# Generate multivariate normal samples for class 1
X1 = multivariate_normal.rvs(mean=mu11, cov=Sigma1, size=round(w1*N1))
X2 = multivariate_normal.rvs(mean=mu12, cov=Sigma1, size=round(w2*N1))
X3 = multivariate_normal.rvs(mean=mu13, cov=Sigma1, size=N1-round(w1*N1)-round(w2*N1))
X = np.vstack((X1, X2, X3))

# Generate samples for class 0
N0 = N - N1
mu0 = np.array([0, 0])
Sigma0 = np.eye(2)
X0 = multivariate_normal.rvs(mean=mu0, cov=Sigma0, size=N0)
X = np.vstack((X, X0))

# Create labels
Y = np.zeros(N)
Y[:N1] = 1

# Create scatter plot
plt.figure()
plt.scatter(X[:N1, 0], X[:N1, 1], c='green', s=10, marker='x', label='green/crosses')
plt.scatter(X[N1:, 0], X[N1:, 1], c='red', marker='o',s=10, label='red/rounds')

plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')
plt.show()
```

## Implementation classifiers


```python
import numpy as np
from scipy.stats import norm

def lin(X, Y):
    N = len(Y)
    X = np.hstack((np.ones((N, 1)), X))
    Yhat = X @ np.linalg.inv(X.T @ X) @ X.T @ Y
    return Yhat

def lda(X, Y):
    sigma = 1  # value fixed in the problem assignment
    I1 = np.where(Y == 1)[0]
    P1 = len(I1) / len(Y)
    muhat1 = np.mean(X[I1, :], axis=0)
    w1 = (muhat1 / sigma**2).reshape(-1, 1)
    w10 = -0.5 / sigma**2 * np.dot(muhat1, muhat1) + np.log(P1)
    return X @ w1 + w10

def knn(X, Y, K=3):
    N = len(Y)
    Yhat = np.zeros(N)
    for i in range(N):
        d = np.sum((X - X[i, :])**2, axis=1)
        nn = np.argsort(d)[:K]
        Yhat[i] = np.mean(Y[nn])
    return Yhat

def nb(X, Y):
    N = len(Y)
    Yhat = np.zeros(N)
    I1 = np.where(Y == 1)[0]
    I0 = np.where(Y == 0)[0]
    p1 = len(I1) / N
    p0 = 1 - p1
    for i in range(N):
        p1x1 = norm.pdf(X[i, 0], np.mean(X[I1, 0]), np.std(X[I1, 0]))
        p1x2 = norm.pdf(X[i, 1], np.mean(X[I1, 1]), np.std(X[I1, 1]))
        
        p0x1 = norm.pdf(X[i, 0], np.mean(X[I0, 0]), np.std(X[I0, 0]))
        p0x2 = norm.pdf(X[i, 1], np.mean(X[I0, 1]), np.std(X[I0, 1]))
        
        Yhat[i] = (p1x1 * p1x2 * p1) / (p1x1 * p1x2 * p1 + p0x1 * p0x2 * p0)
    return Yhat
```

## Computation and plot of ROC curves

```python
# Assuming X and Y are already defined as numpy arrays
Yhat1 = lin(X, Y)
Yhat2 = lda(X, Y)
Yhat3 = nb(X, Y)
Yhat4 = knn(X, Y, K=3)
Yhat5 = knn(X, Y, K=5)
Yhat6 = knn(X, Y, K=10)

# Sort Yhat values in ascending order and get sorted indices
s1 = np.argsort(Yhat1)
s2 = np.argsort(Yhat2)
s3 = np.argsort(Yhat3)
s4 = np.argsort(Yhat4)
s5 = np.argsort(Yhat5)
s6 = np.argsort(Yhat6)

# Initialize True Positive Rate and False Positive Rate lists
TPR1, FPR1 = [], []
TPR2, FPR2 = [], []
TPR3, FPR3 = [], []
TPR4, FPR4 = [], []
TPR5, FPR5 = [], []
TPR6, FPR6 = [], []

# Initialize False Positives and False Negatives lists
FP1, FN1 = [], []
FP2, FN2 = [], []
FP3, FN3 = [], []
FP4, FN4 = [], []
FP5, FN5 = [], []
FP6, FN6 = [], []

N = len(Y)
N1 = np.sum(Y == 1)
N0 = N - N1

for i in range(1, N + 1):
    # LIN
    I1 = s1[:i]
    TPR1.append(np.sum(Y[np.setdiff1d(np.arange(N), I1)] == 1) / N1)
    FPR1.append(np.sum(Y[np.setdiff1d(np.arange(N), I1)] == 0) / N0)
    FP1.append(np.sum(Y[np.setdiff1d(np.arange(N), I1)] == 0))
    FN1.append(np.sum(Y[I1] == 1))
    
    # LDA
    I2 = s2[:i]
    TPR2.append(np.sum(Y[np.setdiff1d(np.arange(N), I2)] == 1) / N1)
    FPR2.append(np.sum(Y[np.setdiff1d(np.arange(N), I2)] == 0) / N0)
    FP2.append(np.sum(Y[np.setdiff1d(np.arange(N), I2)] == 0))
    FN2.append(np.sum(Y[I2] == 1))
    
    # NB
    I3 = s3[:i]
    TPR3.append(np.sum(Y[np.setdiff1d(np.arange(N), I3)] == 1) / N1)
    FPR3.append(np.sum(Y[np.setdiff1d(np.arange(N), I3)] == 0) / N0)
    FP3.append(np.sum(Y[np.setdiff1d(np.arange(N), I3)] == 0))
    FN3.append(np.sum(Y[I3] == 1))
    
    # 3NN
    I4 = s4[:i]
    TPR4.append(np.sum(Y[np.setdiff1d(np.arange(N), I4)] == 1) / N1)
    FPR4.append(np.sum(Y[np.setdiff1d(np.arange(N), I4)] == 0) / N0)
    FP4.append(np.sum(Y[np.setdiff1d(np.arange(N), I4)] == 0))
    FN4.append(np.sum(Y[I4] == 1))
    
    # 5NN
    I5 = s5[:i]
    TPR5.append(np.sum(Y[np.setdiff1d(np.arange(N), I5)] == 1) / N1)
    FPR5.append(np.sum(Y[np.setdiff1d(np.arange(N), I5)] == 0) / N0)
    FP5.append(np.sum(Y[np.setdiff1d(np.arange(N), I5)] == 0))
    FN5.append(np.sum(Y[I5] == 1))
    
    # 10NN
    I6 = s6[:i]
    TPR6.append(np.sum(Y[np.setdiff1d(np.arange(N), I6)] == 1) / N1)
    FPR6.append(np.sum(Y[np.setdiff1d(np.arange(N), I6)] == 0) / N0)
    FP6.append(np.sum(Y[np.setdiff1d(np.arange(N), I6)] == 0))
    FN6.append(np.sum(Y[I6] == 1))

# Plot ROC curves
plt.figure(figsize=(8, 6))
plt.plot(FPR1, TPR1, color='yellow', label='LIN')
plt.plot([0, 1], [0, 1], linestyle='--', color='grey')
plt.plot(FPR2, TPR2, color='black', label='LDA')
plt.plot(FPR3, TPR3, color='orange', label='NB')
plt.plot(FPR4, TPR4, color='blue', label='3NN')
plt.plot(FPR5, TPR5, color='magenta', label='5NN')
plt.plot(FPR6, TPR6, color='green', label='10NN')

plt.title('ROC Curve')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend(fontsize=8, loc='lower right', ncol=3)
plt.show()
```

Looking at the ROC curves the best two classifiers are the 3NN and 5NN.
For a better assessment of their accuracy the area under the ROC curve should be computed.
